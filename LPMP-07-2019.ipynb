# Licitaciones Mercado Público a Julio 2019

Archivo Mercado Público Julio 2019

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import pandas as pd
import pandas_profiling
df = pd.read_csv('data/LPMP-07-2019.csv')
df.shape
```

```python
#df.drop(["Textbox40"],axis=1, inplace=True)
df.head()
```

```python
df.columns
```

```python
df.groupby(('Textbox39','FechaCierre1')).count()['Textbox36']
```

```python
# Visualiza estos valores en un diagrama de barras.
ax = df.groupby('Textbox39').count()['Textbox36'].plot(
kind='barh',
color='orange',
grid=True,
figsize=(12, 120),
title='Agrupado por Publicación Institucion '
)
ax.set_xlabel('Institucion Publica')
ax.set_ylabel('Licitaciones Mercado Público Julio 2019')
plt.show()
```
