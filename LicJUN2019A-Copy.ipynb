```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('data/Licitacion_Publicada.csv', encoding='latin-1', low_memory=False)
df.shape
```

```python
df.head()
```

```python
df.columns
```

```python
df.groupby('TipoLC').count()['Textbox36']
```

```python
# Visualiza estos valores en un diagrama de barras.
ax = df.groupby('TipoLC').count()['Textbox36'].plot(
kind='bar',
color='orange',
grid=True,
figsize=(15, 7),
title='Agrupado por Tipo Licitación'
)
ax.set_xlabel('Tipo Licitación')
ax.set_ylabel('Total de Licitaciones')
plt.show()
```

Agrupamos x Region solo Licitaciones Públicas Mayor a 5000 UTM

```python
by_df = df.loc[df['TipoLC'] == 'Licitacion Publica Mayor a 5000 UTM', ].groupby(['citName']).count()['FechaCierre1']
by_df.to_csv('data/regiones.csv', index=True)
```

```python
by_df
```

```python
# Visualiza en un gráfico de barras horizontales
ax = (by_df).plot(
kind='barh',
figsize=(10, 7),
color='palegreen',
grid=True
)
ax.set_xlabel('Distribucion x Region (Licitacion Pública Mayor a 5000 UTM)')
plt.show()
```
