   Datos del Mercado Público https://www.mercadopublico.cl/Home

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('data/Licitacion_Publicada.csv', encoding='latin-1', low_memory=False)
df.head()
```


Agrupamos x Region solo Licitacion Publica Mayor a 5000 UTM


```python
by_offer_type_df = df.loc[
df['TipoLC'] == 'Licitacion Publica Mayor a 5000 UTM' # clientes comprometidos
].groupby([
'TipoLC', 'citName' # Agrupando los datos por estas dos columnas.
]).count()['FechaCierre1'] #/ df.groupby('TipoLC').count()['FechaCierre1']#
by_offer_type_df
```

```python
by_offer_type_df = by_offer_type_df.unstack().fillna(0)
by_offer_type_df
```

```python
# Visualizar en gráfico de barras
ax = (by_offer_type_df*5*4.7/70).plot(
kind='bar',
figsize=(20, 7),
grid=True
)
ax.set_ylabel('N° Licitaciones Agrupadas x Region Total en Millones de US$')
plt.show()
```
